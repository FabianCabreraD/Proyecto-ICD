```python
import matplotlib.pyplot as plt
import csv
```

```python
with open("d:\\uh\\icd\\Proyecto-ICD\\data\\tasa copy.csv","r") as file:
    data = csv.reader(file)
    data_list = [i for i in data]

data_list = data_list[1:]
data_list_final = []
for i in data_list:
    row = [i[0],float(i[1]),float(i[2])]
    data_list_final.append(row)
```

```python
fechas = [i[0] for i in data_list_final]
usd = [i[1] for i in data_list_final]
euro = [i[2] for i in data_list_final]
```

```python
plt.figure()
plt.plot(fechas,usd,label="USD")
plt.plot(fechas,euro,label="EUR")
plt.xticks(fechas[::15],rotation=45)
plt.title("Tasa cambiaria en los Ãºltimos 3 meses")
plt.xlabel("fecha")
plt.ylabel("cambio en CUP")
plt.legend()
plt.show()
```

```python
fechas = [i[0] for i in data_list_final]
usd = [i[1] for i in data_list_final]
euro = [i[2] for i in data_list_final]
```

```python
usd_vs_salario = [6660.1/i for i in usd]
euro_vs_salario = [6660.1/i for i in euro]
```

```python
plt.figure()
plt.plot(fechas,usd_vs_salario,label="USD")
plt.plot(fechas,euro_vs_salario,label="EUR")
plt.xticks(fechas[::15],rotation=45)
plt.title("Salario Medio en Cuba en USD y EUR")
plt.xlabel("fecha")
plt.legend()
plt.show()
```

```python
usd_vs_jubilacion_minima = [3056/i for i in usd]
euro_vs_jubilacion_minima = [3056/i for i in euro]
```

```python
plt.figure()
plt.plot(fechas,usd_vs_jubilacion_minima,label="USD")
plt.plot(fechas,euro_vs_jubilacion_minima,label="EUR")
plt.xticks(fechas[::15],rotation=45)
plt.title("JubilaciÃ³n MÃ­nima en Cuba en USD y EUR")
plt.xlabel("fecha")
plt.legend()
plt.show()
```

```python
#Salario y jubilacion
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(12,6))

ax1.plot(fechas,usd_vs_salario,label="USD")
ax1.plot(fechas,euro_vs_salario,label="EUR")
ax1.set_xticks(fechas[::15])
ax1.set_xlabel("fecha")
ax1.tick_params(axis="x",rotation=45)
ax1.legend()
ax1.set_title("Salario Promedio en USD y EUR")

ax2.plot(fechas,usd_vs_jubilacion_minima, label="USD")
ax2.plot(fechas,euro_vs_jubilacion_minima, label="EUR")
ax2.set_xticks(fechas[::15])
ax2.set_xlabel("fecha")
ax2.tick_params(axis="x",rotation=45)
ax2.legend()
ax2.set_title("PensiÃ³n mÃ­nima en USD y EUR")

plt.show()
```
