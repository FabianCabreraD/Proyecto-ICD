```python
import icd_module as icd
import matplotlib.pyplot as plt 
```

```python
precios_cerveza_nacional = []
precios_cerveza_importada = []
archivos = icd.mipymes_list()
for i in archivos:
    path = icd.return_path_mipymes(i)
    file_mipyme = icd.read_json(path)
    
    productos = file_mipyme["product"]
    
    for j in productos:
        if j["type"] == "Cerveza":
            if j["origin"] == "Nacional":
                precios_cerveza_nacional.append(j["price"])
            elif j["origin"] == "Importado":
                precios_cerveza_importada.append(j["price"])
        
promedio_cn = icd.mean_list(precios_cerveza_nacional)
promedio_ci = icd.mean_list(precios_cerveza_importada)

promedio_cn, promedio_ci

data = {
    "Nacional": promedio_cn,
    "Importado": promedio_ci
}
    
```

```python
fig, ax = plt.subplots()

ax.bar(data.keys(),data.values())
ax.set_ylabel("Precio en CUP")
ax.set_xlabel("Origen")
ax.set_ylim(200,270)
plt.show()
```
